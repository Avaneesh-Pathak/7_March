Q1. The three measures of central tendency are the mean, median, and mode.

Q2. The mean is the average value of a dataset, calculated by summing up all the values and dividing by the total number of values. The median is the middle value of a dataset when it is arranged in ascending or descending order. If the dataset has an even number of values, the median is the average of the two middle values. The mode is the value that appears most frequently in the dataset.

These measures are used to determine the typical or central value of a dataset. The mean provides an overall average, the median represents the middle value, and the mode shows the most common value. They give different perspectives on the central tendency of a dataset and can help identify any outliers or unusual patterns.

Q5. Measures of dispersion, such as range, variance, and standard deviation, provide information about the spread or variability of a dataset.

- Range: It is the simplest measure of dispersion and represents the difference between the maximum and minimum values in a dataset. It gives an idea of the overall spread but does not consider the distribution of values within the range.

- Variance: It measures the average squared deviation from the mean. It considers the spread of all values in the dataset and provides a measure of how much the values deviate from the mean.

- Standard Deviation: It is the square root of the variance and represents the average deviation from the mean. It provides a measure of the spread while using the same unit of measurement as the original data.

For example, let's consider the dataset [10, 15, 20, 25, 30]. The range is 30 - 10 = 20. The variance is calculated as [(10-20)² + (15-20)² + (20-20)² + (25-20)² + (30-20)²] / 5 = 50. The standard deviation is the square root of the variance, so it is √50 ≈ 7.07. These measures describe the spread of the dataset, with the standard deviation giving a more easily interpretable measure of the average deviation from the mean.

Q8. Skewness in data refers to the measure of asymmetry or departure from symmetry in a dataset's distribution. It indicates whether the data is symmetrically distributed or skewed to one side. Skewness can be positive, negative, or zero.

- Positive skewness: The tail of the distribution extends towards the right side, and the majority of the data is concentrated on the left side. The mean is typically greater than the median in positively skewed data.

- Negative skewness: The tail of the distribution extends towards the left side, and the majority of the data is concentrated on the right side. The mean is typically less than the median in negatively skewed data.

- Zero skewness: The distribution is symmetric, with equal concentrations of data on both sides of the mean. The mean and median are approximately equal in data with zero skewness.

Q9. If a dataset is right skewed, the median will generally be less than the mean. This is because the skewed tail on the right side (higher values) pulls the mean towards that direction, making it greater than the median. The median represents the central value of the dataset, whereas the mean is affected by extreme values in the tail.

Q10. Covariance and correlation are both measures used to describe the relationship between two variables in statistical analysis.

- Covariance: It measures the extent to which two variables vary together

in a dataset. Covariance can be positive, indicating a direct relationship where both variables increase or decrease together, or negative, indicating an inverse relationship where one variable increases while the other decreases. However, covariance does not provide a standardized measure, making it difficult to compare across different datasets.

- Correlation: It is a standardized measure that quantifies the strength and direction of the linear relationship between two variables. The correlation coefficient, typically denoted by "r," ranges from -1 to 1. A correlation coefficient of 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no linear correlation. Correlation takes into account the scales of the variables and provides a more interpretable measure of the relationship compared to covariance.

Both covariance and correlation are used to analyze the relationship between variables. Covariance helps determine whether the variables tend to move in the same or opposite direction, while correlation additionally measures the strength and direction of the linear association. These measures are utilized in various statistical analyses, such as regression analysis, portfolio management, and determining the strength of relationships in scientific studies.

Q11. The formula for calculating the sample mean is:

Sample Mean (x̄) = (x₁ + x₂ + ... + xn) / n

where x₁, x₂, ..., xn are the individual values in the dataset, and n is the total number of values.

For example, let's calculate the sample mean for the dataset [5, 10, 15, 20]:

Sample Mean (x̄) = (5 + 10 + 15 + 20) / 4 = 12.5

Therefore, the sample mean for the given dataset is 12.5.

Q12. For a normal distribution, the measures of central tendency (mean, median, and mode) are equal. In other words, the mean, median, and mode all have the same value, which is the peak or center of the normal distribution. The normal distribution is symmetric, so the data is evenly distributed around the mean, resulting in the equality of these measures.

Q14. Outliers can significantly affect measures of central tendency and dispersion. An outlier is an extreme value that is distant from other values in a dataset. Here's an example:

Consider the dataset: [10, 20, 30, 40, 50, 1000]

- Measures of central tendency: The mean is heavily influenced by outliers. In this case, the mean would be (10 + 20 + 30 + 40 + 50 + 1000) / 6 ≈ 185. The presence of the outlier, 1000, significantly increases the mean, making it an unreliable representation of the typical value. The median, on the other hand, is not as affected by outliers and remains at 35, which is a better indicator of the central value.

- Measures of dispersion: Outliers can greatly impact measures of dispersion, such as the range, variance, and standard deviation. The range, which is the difference between the maximum and minimum values, would be 1000 - 10 = 990, mainly due to the outlier. The variance and standard deviation would also be inflated by the presence of the outlier, indicating a higher spread or variability in the data.

In this example, the outlier significantly affects both measures of central tendency and dispersion, highlighting the importance of identifying and dealing with outliers appropriately when analyzing data.
